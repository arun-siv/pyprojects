```python
sales_data = pd.read_csv("sales_data.csv")
sales_data
```

```python
sales_data.date_time = pd.to_datetime(sales_data['date_time'])

type(sales_data.date_time[0])
```

```python
import datetime
import random
delivery_dates = []

for dt in sales_data.date_time:
   delivery_dates.append(dt + datetime.timedelta(days=random.randint(1,7)))
```

```python
delivery_dates
```

```python
sales_data["delivery_dates"] = delivery_dates
sales_data
```

```python
sales_data["time_delta"] = sales_data["delivery_dates"] - sales_data["date_time"]
sales_data
```

```python
sales_data["product"] = pd.Categorical(sales_data["product"])
sales_data["location"] = pd.Categorical(sales_data["location"])
```

```python
sales_data["product"]
```

```python
sales_data["product"] = sales_data["product"].cat.remove_categories("Bag")
```

```python
sales_data.dropna()
```

```python
sales_data.groupby("location")["time_delta"].mean()
```

```python
sales_data.groupby("product")["time_delta"].mean()
```
